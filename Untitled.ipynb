# 1) Lorenz system

```python
import numpy as np
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from functools import partial
from scipy.optimize import minimize as opt
import scipy.integrate
from sklearn import linear_model
import pylab
```

```python
def lorenz(state, t, p):
    x, y, z = state
    sigma, beta, rho = p
    
    dx = sigma * (y - x)
    dy = x * (rho - z) - y
    dz = x * y - beta * z
     
    return np.array([dx, dy, dz])
```

```python
sigma = 10.0
beta = 8.0 / 3.0
rho = 28.0

p = (sigma, beta, rho)
```

```python
y0 = [1.0, 1.0, 1.0]

t = np.arange(0.0, 30.0, 0.01)
 
result = odeint(lorenz, y0, t, (p, ))
pure_sys = result.copy()
    
fig = plt.figure(figsize=(10, 8))
ax = fig.gca(projection="3d")
ax.plot(result[:, 0], result[:, 1], result[:, 2])
```

```python
X = result
X_dot = []
for i, val in enumerate(X):
    X_dot.append(lorenz(val, '', p) + np.random.normal(scale = 0.5, size = 3))
    X[i] += np.random.normal(scale = 0.3, size = 3)
X_dot = np.array(X_dot)
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.gca(projection="3d")
ax.plot(X[:, 0], X[:, 1], X[:, 2], color = 'b')
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.gca(projection="3d")
ax.plot(X_dot[:, 0], X_dot[:, 1], X_dot[:, 2], color = 'b')
```

```python
def features(state, t):
    features = []
    features.extend(state)
    for i in range(len(state)):
        for j in range(i, len(state)):
            features.append(state[i]*state[j])
    return np.array(features)
```

```python
def solver(coef_mat, reg, x_dot, num):
    temp_mat = coef_mat
    struct = [1 for _ in range(coef_mat.shape[1])]
    res = 0
    for _ in range(num):
        reg.fit(temp_mat, x_dot)
        res = reg.coef_
        counter = -1
        for j in range(len(struct)):
            if struct[j] != 0:
                counter+=1
            if counter == np.argmin(np.abs(res)):
                struct[j] = 0
                break
        temp_mat = np.delete(temp_mat, np.argmin(np.abs(res)), axis=1)
    reg.fit(temp_mat, x_dot)
    res = reg.coef_ 
    i, j = 0, 0
    while i<len(struct):
        if struct[i] != 0:
            struct[i] = struct[i]*res[j]
            j+=1
        i+=1
    return struct
```

## 3.2) Sequential thresholded least-squares

```python
cutoff = 0.2
```

```python
def solver_hard(coef_mat, reg, x_dot, cutoff):
    temp_mat = coef_mat
    struct = [1 for _ in range(coef_mat.shape[1])]
    res = 0
    prev = -1
    while 1:
        reg.fit(temp_mat, x_dot)
        res = reg.coef_
        counter = -1
        if prev == len([1 for i in res if abs(i)>=cutoff]):
            break
        zero_list = []
        for i, val in enumerate(res):
            if abs(val)<cutoff:
                zero_list.append(i)
        for i in reversed(zero_list):
            temp_mat = np.delete(temp_mat, i, axis=1)
        counter = -1
        for j in range(len(struct)):
            if struct[j] != 0:
                counter+=1
            if counter in zero_list:
                struct[j] = 0
        prev = len([1 for i in struct if i!=0])
    reg.fit(temp_mat, x_dot)
    res = reg.coef_ 
    i, j = 0, 0
    while i<len(struct):
        if struct[i] != 0:
            struct[i] = struct[i]*res[j]
            j+=1
        i+=1
    
    return struct
```

# 2) LotkaVolterra equation

```python
def LVe(state, t, p):
    x, y = state
    alpha, beta, delta, gamma = p
    
    dx = x*(alpha - beta*y)
    dy = y*(-delta + gamma*x)
     
    return np.array([dx, dy])
```

```python
# alpha = 1.
# beta = 1.5
# delta = 0.7
# gamma = 1.2
alpha = 3.0
beta = 2.3
delta = 0.8
gamma = 1.0

p = (alpha, beta, delta, gamma)
```

```python
y0 = [0.5, 0.1]
step = 0.01
t = np.arange(0.0, 15.0, step)
 
result = odeint(LVe, y0, t, (p, ))
pure_sys = result.copy()
    
pylab.plot(result[:, 0], result[:, 1])    
```
